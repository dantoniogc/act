```python
# Cargar librerías
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay

print("Librerías cargadas correctamente.")
```

```python
# Cargar datos
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names
target_names = data.target_names

print(f"Dataset cargado: {len(X)} muestras y {len(feature_names)} características.")
```

```python
# Preprocesamiento de datos
# División 80% entrenamiento y 20% prueba
train_data, test_data, train_labels, test_labels = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(f"Datos de entrenamiento: {len(train_data)}")
print(f"Datos de prueba: {len(test_data)}")
```

```python
# Selección y entrenamiento del modelo
modelo = DecisionTreeClassifier(random_state=42)

# Entrenamiento
modelo.fit(train_data, train_labels)

print("El modelo de árbol de decisión ha sido entrenado.")
```

```python
# Prueba y metricas

# Prueba del modelo
predicciones = modelo.predict(test_data)

# Métricas de precisión
accuracy = accuracy_score(test_labels, predicciones)
print(f"--- PRECISIÓN (ACCURACY): {accuracy:.4f} ---\n")

# Reporte de clasificación detallado
print("--- REPORTE DE CLASIFICACIÓN ---")
print(classification_report(test_labels, predicciones, target_names=target_names))

# Matriz de Confusión visual
print("\n--- MATRIZ DE CONFUSIÓN ---")
conf_matrix = confusion_matrix(test_labels, predicciones)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
# Representación gráfica
print("Generando el gráfico del árbol de decisión (esto puede tardar unos segundos por el DPI)...")

plt.figure(figsize=(25,12), dpi=300)
plot_tree(modelo,
          feature_names=feature_names,
          class_names=target_names,
          filled=True,
          rounded=True,
          fontsize=10)

plt.title("Visualización del Árbol de Decisión - Breast Cancer Dataset", fontsize=20)
plt.show()
```
